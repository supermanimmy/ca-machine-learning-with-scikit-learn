```python
import pandas as pd

df = pd.read_csv('data/restaurant.csv')
```

```python
df.head()
```

```python
#One hot encoding
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
ohe_df = pd.DataFrame(ohe.fit_transform(df).toarray())
```

```python
ohe_df.head()
```

```python
ohe = OneHotEncoder()
pd.DataFrame(ohe.fit_transform(df[['smoker']]).toarray()).head()
```

```python
# dummy variables, they can take the value of 0 or 1
```

```python
ohe.categories_
```

```python
from sklearn.compose import make_column_transformer
```

```python
ohe = OneHotEncoder()
ct = make_column_transformer(
    (ohe, ['smoker', 'sex', 'day', 'time']),
    remainder='passthrough'
)
```

```python
data_new = pd.DataFrame(ct.fit_transform(df))
```

```python
data_new.head()
```

```python
#toy dataset
data_toy = pd.DataFrame({
    'constructor': ['Toyota', 'Audi', 'Renault']
})
```

```python
ohe_toy  = OneHotEncoder()
pd.DataFrame(ohe_toy.fit_transform(data_toy).toarray())
```

```python
new_data = pd.DataFrame(
{
    'constructor': ['Toyota', 'Jeep']
})
```

```python
ohe_toy.transform(new_data)
```

```python
ohe_toy  = OneHotEncoder(handle_unknown='ignore', sparse=False)
pd.DataFrame(ohe_toy.fit_transform(data_toy))
```

```python
pd.DataFrame(ohe_toy.transform(new_data))
```
